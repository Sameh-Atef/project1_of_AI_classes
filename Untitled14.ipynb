```python
import pandas as pd 
dataset=pd.read_csv(r"/home/sameh/Desktop/task3/Wuzzuf_Jobs.csv")
dataset.head()
```

```python
dataset.describe()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#print(dataset.isnull().sum())
#print(sum(dataset.duplicated()))
dataset.drop_duplicates(inplace =True)
print(sum(dataset.duplicated()))
```

```python
companies = dataset['Company'].value_counts()
companies
```

```python
c= companies[:10].index
values = companies[:10].values
plt.pie(values ,labels =c)
plt.legend()
plt.figure
plt.show();
```

```python
jobs =dataset["Title"].value_counts()
jobs
```

```python
j = jobs[:10].index
values = companies[:10].values
plt.pie(values ,labels =j)
plt.legend()
plt.figure
plt.show();
```

```python
areas =dataset["Location"].value_counts()
areas
```

```python
A = areas.index[:10]
values = areas.values [:10]
plt.bar (A ,values )
plt.title("The Most Popular Areas" ,fontsize=25)
plt.xlabel("Areas" ,fontsize =20)
plt.ylabel ("Number" ,fontsize =20)
plt.xticks(rotation=90)
plt.show()
```

```python
skill =dataset["Skills"].str.cat(sep =",")
skills_count =pd.Series (skill.split(",")).value_counts()

print (skills_count)
```

```python
S = skills_count.index[:10]
values = skills_count.values [:10]
plt.bar (S ,values )
plt.title("The Most Popular Skills" ,fontsize=20)
plt.xlabel("Skill" ,fontsize =25)
plt.ylabel ("How many each repeated" ,fontsize =15)
plt.xticks(rotation=90)
plt.show()
```
